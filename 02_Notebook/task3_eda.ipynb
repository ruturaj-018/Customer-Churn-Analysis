```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
file_path = "cleaned_data.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
df.info()
```

```python
plt.figure()
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.xlabel("Churn (0 = No, 1 = Yes)")
plt.ylabel("Customer Count")
plt.show()
```

```python
plt.figure()
sns.countplot(x='gender', hue='Churn', data=df)
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure()
sns.countplot(x='SeniorCitizen', hue='Churn', data=df)
plt.title("Churn by Senior Citizen")
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Churn by Contract Type")
plt.savefig("charts/churn_by_contract.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='tenure', hue='Churn', bins=30, kde=True)
plt.title("Tenure Distribution by Churn")
plt.savefig("charts/churn_by_tenure.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.savefig("charts/churn_by_monthly_charges.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y='PaymentMethod', hue='Churn', data=df)
plt.title("Churn by Payment Method")
plt.savefig("charts/churn_by_payment_method.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title("Churn by Internet Service")
plt.savefig("charts/churn_ by_Internet_Service.png", bbox_inches='tight')
plt.show()
```
